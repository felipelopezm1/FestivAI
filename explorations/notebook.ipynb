```python
import pandas as pd
import json
```

```python
#Thursday
with open("lollapalooza_thursday_2025.json", "r") as file:
    thursday_schedule = json.load(file)

data = []
for stage, performances in thursday_schedule.items():
    for performance in performances:
        data.append({
            "Stage": stage,
            "Artist": performance["artist"],
            "Time": performance["time"]
        })

df = pd.DataFrame(data)

print(df)
```
